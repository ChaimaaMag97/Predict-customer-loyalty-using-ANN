# The imports

```python
import pandas as pd
import keras
```

```python
data=pd.read_csv("Churn_Modelling.csv")
```

```python
data.head()
```

```python
a=pd.get_dummies(data, columns=['Geography'])
```

```python
data=pd.get_dummies(a, columns=['Gender'])
```

```python
data
```

```python
data.columns
```

```python
data=data.drop(['RowNumber', 'CustomerId', 'Geography_Spain','Gender_Female','Surname'], axis=1)
data
```

```python
data=data[[c for c in data if c not in ['Exited']]  + ['Exited']]
data
```

```python
X = data.iloc[:, 1:11].values
y = data.iloc[:, 11].values
```

```python
y
```

### splitting the dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

### Feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

### creation of artifical neurol network

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
classifier1=Sequential()
```

```python
classifier1.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_shape= (10,)))

```

```python
classifier1.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
```

```python
classifier1.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
classifier1.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
classifier1.fit(X_train,y_train,batch_size=10,epochs=100)
```

```python
y_pred=classifier1.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
y_pred = (y_pred > 0.5)
```

```python
cm=confusion_matrix(y_test,y_pred)
```
